<h1 style="font-size:30px">Plot Section</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1.1
def sinfunc(x,A,w):
	return A*np.sin(w*x)
```

```python
#1.2
x=np.linspace(0,2*np.pi,1000)
```

```python
#1.3
A=np.array([0.5,1,1.5,2,2.5])
w=np.array([1,2,3,4,5])
```

```python
fw=list(zip(A,w))
print((fw[0:1]))
#1.4
plt.figure()
```

```python
#1.5
for i in range(int(5)):
    A,w=(fw[i])
    y=sinfunc(x,float(A),float(w))
    plt.plot(x,y)
    plt.legend(fw)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Sin functions for different amplitudes ")
plt.show()
```
